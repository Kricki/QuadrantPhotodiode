# Basic example

Create QPD object. `size`, `gap` and `shape` define the geometry of the detector.

The attribute `laser_intensity` stores the beam profile (elliptic Gaussian beam in this example).

```python
qpd = quadrantphotodiode.QPD(size=5e-3, gap=40e-6, shape='circular')
x, y = qpd.xy_meshgrid()
qpd.laser_intensity = phytools.functions.gaussian2d(x=x, y=y, a=1, x0=-0.4e-3, y0=0.6e-3, fwhm_x=1.8e-3, fwhm_y=2.5e-3, offset=0)
```

Plot the active region of the QPD.

```python
qpd.plot('detector_active')
```

Plot the laser beam profile.

```python
qpd.plot('laser_intensity')
```

Plot the intensity as measured by the QPD.

```python
qpd.plot('detector_intensity')
```

Compute the intensity inside the individual quadrants:

```python
q1, q2, q3, q4 = qpd.quadrants
print('Quadrant 1: ' + str(q1))
print('Quadrant 2: ' + str(q2))
print('Quadrant 3: ' + str(q3))
print('Quadrant 4: ' + str(q4))
```

Compute the x and y position (in arbitrary units):

```python
print('x_diff: ' + str(qpd.x_diff))
print('y_diff: ' + str(qpd.y_diff))
print('x_pos: ' + str(qpd.x_pos))
print('y_pos: ' + str(qpd.y_pos ))
```
